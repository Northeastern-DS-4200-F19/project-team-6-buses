```python
import pandas as pd
import glob

path = 'data/OTP by month' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True, sort=True)

frame.drop(columns='Unnamed: 12')
```

```python
#filters

filter1 = frame['gtfs_route_id'] == "1"

filter2 = frame['gtfs_route_id'] == "43"

filter3 = frame['gtfs_route_id'] == "749"

filter4 = frame['gtfs_route_id'] == "751"
```

```python
frame.where(filter1 | filter2 | filter3 | filter4, inplace = True)

frame_ColumnsNeeded = frame[["gtfs_route_id","service_date","peak_offpeak_ind","otp_numerator","otp_denominator"]].dropna()

frameFinal = (frame_ColumnsNeeded[pd.notnull(frame_ColumnsNeeded[["gtfs_route_id","service_date","peak_offpeak_ind","otp_numerator","otp_denominator"]])]).sort_values(by=['service_date']).reset_index().drop(columns="index")
```
