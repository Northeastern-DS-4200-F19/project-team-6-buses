```python
#Route 1 line with its coinciding shapeID (one direction)
filter1 = shapes['shape_id']=='010058'

#Route 43 with its coinciding shapeID (one direction)
filter2 = shapes['shape_id']=='430031'

#Route SL5 with its coinciding shapeID (one direction)
filter3 = shapes['shape_id']=='7490022'

#Route SL4 with its coinciding shapeID (one direction)
filter4 = shapes['shape_id']=='7510024'

#Route 1 line with another coinciding shapeID (second direction)
filter5 = shapes['shape_id']=='010070'

#Route 43 line with another coinciding shapeID (second direction)
filter6 = shapes['shape_id']=='430029'

#Route SL5 line with another coinciding shapeID (second direction)
filter7 = shapes['shape_id']=='7490023'

#Route SL4 line with another coinciding shapeID (second direction)
filter8 = shapes['shape_id']=='7510025'
```
