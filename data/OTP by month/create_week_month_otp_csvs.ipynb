```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('Full_List_of_ChesterSquare_stops_OTPs.csv')

def str_to_date(string):
    return datetime.strptime(string, '%Y-%m-%d')

df["service_date"] = df["service_date"].apply(str_to_date)
```

```python
df.head()
```

```python
df['month'] = df["service_date"].dt.month
```

```python
df = df.drop(columns=["Unnamed: 0", "otp_numerator", "otp_denominator"])
df5 = df.groupby(["month", "gtfs_route_id"]).mean()
df5
```

```python
df5.to_csv("Month-by-Month OTP Score.csv")
```
